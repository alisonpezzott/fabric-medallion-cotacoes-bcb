```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
import requests
import json
```

```python
url = "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/Moedas?$top=100&$format=json&$select=simbolo,nomeFormatado"

response = requests.get(url).json()['value']

df = spark.createDataFrame(response).selectExpr(
    "simbolo AS Moeda",
    "nomeFormatado AS MoedaNome"
)

df.write.saveAsTable('dim_moedas')
```

```python
%%sql
SELECT * FROM dim_moedas
```
